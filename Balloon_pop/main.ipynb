```python
# Importing required libraries
import random
import pygame # open-source library for the development of multimedia apps.
import cv2 
import numpy as np
from cvzone.HandTrackingModule import HandDetector # To detect hands
import time

# Initializing the pygame
pygame.init() # for initialize all imported pygame modules 

# Creating Window 
width, height = 1280, 720 # set width and height of window
window = pygame.display.set_mode((width, height)) # set window
pygame.display.set_caption("Balloon PoP") # set caption

# Initializing Clock for FPS
fps = 30 # set fps (frame per second)
clock = pygame.time.Clock() # create a object for clock used to control the game framerate

# Opening Webcam 
cam = cv2.VideoCapture(0) # to open defualt camara
cam.set(3, 1280)  # set width
cam.set(4, 720)  # set height

# Importing Image of balloon and create a rect
img_balloon = pygame.image.load(r'D:\DL Projects\first_project\Balloon_pop\Assets\BalloonRed.png').convert_alpha() # set balloon image
rect_balloon = img_balloon.get_rect() # create a rectangle 
rect_balloon.x, rect_balloon.y = 500, 300 # set x and y values

# Importing pop sound
pop_sound=pygame.mixer.Sound(r"D:\DL Projects\first_project\Balloon_pop\Assets\pop2-84862.wav")

# Settings Variables
speed = 20
score = 0
startTime = time.time()
totalTime = 30

# Importing hand detector in a variable
detector = HandDetector(detectionCon=0.8, maxHands=2) # to detect hands

# placing balloon at random position
def reset_balloon():
    rect_balloon.x = random.randint(100, img.shape[1] - 100) 
    rect_balloon.y = img.shape[0] + 50

# Main loop
start = True
while start:
    # when quit the game
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            start = False
            pygame.quit()
    # When the time is end
    time_remain = int(totalTime - (time.time() - startTime)) # set time runner
    if time_remain < 0:
        window.fill((255, 255, 255)) # window filled with white col0ur
        font = pygame.font.Font(r'D:\DL Projects\first_project\Balloon_pop\Assets\Marcellus-Regular.ttf', 50) # set font
        text_score = font.render(f'Your Score: {score}', True, (0, 0, 255)) # set score bord
        text_time = font.render(f'Time UP', True, (0, 0, 255))  
        window.blit(text_score, (450, 350))
        window.blit(text_time, (530, 275))
    # The game code
    else:
        success, img = cam.read() 
        img = cv2.flip(img, 1) # capture frame by frame
        hands, img = detector.findHands(img, flipType=False) # find hands
        # Move the balloon up
        rect_balloon.y -= speed 
        # if balloon has reached the top without pop
        if rect_balloon.y < 0:
            reset_balloon()
            speed += 1 # increase speed +1 
        # hand detection
        if hands:
            hand = hands[0]
            x, y = hand['lmList'][8][0:2] # to access the tip of index fingers
            # if the balloon has poped
            if rect_balloon.collidepoint(x, y): # it is a method of the Rect class. It is used to test if a point is inside a rectangle
                pop_sound.play() # play the pop sound
                reset_balloon()
                score += 5 # increase score +5
                speed += 1 # increase speed +1
        # set window
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert bgr to rgb 
        imgRGB = np.rot90(imgRGB) # rotate 90 degrees counterclockwise
        frame = pygame.surfarray.make_surface(imgRGB).convert() # to make a pygame surface
        frame = pygame.transform.flip(frame, True, False)  # to flips the pygame surface horizontally
        window.blit(frame, (0, 0)) # to draw the frame surface onto the window surface at the top-left corner.
        window.blit(img_balloon, rect_balloon) # To draw the img_balloon onto the window surface at the position specified by rect_balloon
        font = pygame.font.Font(r'D:\DL Projects\first_project\Balloon_pop\Assets\Marcellus-Regular.ttf', 50) # set font
        text_score = font.render(f'Score: {score}', True, (50, 50, 255)) # set score
        text_time = font.render(f'Time: {time_remain}', True, (50, 50, 255)) # set time
        window.blit(text_score, (35, 35))
        window.blit(text_time, (1000, 35))

    # Update Display
    pygame.display.update()
    
    # Set FPS
    clock.tick(fps) # if game loop runs too quicklty, this will pause the game loop to ensure it doesnpt exceed the fps



```
